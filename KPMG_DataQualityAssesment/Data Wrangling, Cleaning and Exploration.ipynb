**Data Cleaning**

Issues :

**Drop the unnecesary columns from the datasets.**

```python
#dropping the country column from the customer Adrress all customers are from the same country
CustomerAddress1 = CustomerAddress1.drop(['country'],axis = 1)
```

```python
#confirming the column has been dropped succefully
CustomerAddress1.info()
```

```python
#columns dropped from the customer demoghrapic;DOB,Default.
CustomerDemographic1 =CustomerDemographic1.drop(['DOB','deceased_indicator','default'],axis =1)
```

```python
#confirming the columns were dropped succefully
CustomerDemographic1.info()
```

```python
Transactions1 =Transactions1.drop(['transaction_date','list_price','standard_cost','product_first_sold_date'],axis =1)
```

```python
Transactions1.info()
```

Join the datasets together to address the untidy nature of them being separate.

```python
#joining CustomerAddress1 and CustomerDemographic1 so that it retains only Customers information in both datasets.
Customer_info = CustomerAddress1.merge(CustomerDemographic1, on='customer_id')
Customer_info.head()
```

```python
#joining Customer_info and Transactions1 so that it retains only Customers with transactions.
Customer_Transactions = Transactions1.merge(Customer_info, on='customer_id')
Customer_Transactions.head()
```

```python
Customer_Transactions.shape
```

```python
Customer_Transactions.info()
```

```python
Customer_Transactions.isna().sum()
```

```python
Customer_Transactions.duplicated().sum()
```

```python
#Dropping the null values
Customer_Transactions =Customer_Transactions.dropna()
```

```python
#In Gender column replace F and Femal with Female
Customer_Transactions['gender'] = Customer_Transactions['gender'].replace(['F','Femal'], 'Female')
```

```python
#confirming the replacement was successful through condional extraction
F = Customer_Transactions['gender'] == 'F'
print(F)
```

```python
Femal = Customer_Transactions['gender'] == 'Femal'
print(Femal)
```

```python
#In Gender column replace M with Male
Customer_Transactions['gender'] = Customer_Transactions['gender'].replace(['M'], 'Male')
```

```python
#confirming the replacement was successful through condional extraction
M = Customer_Transactions['gender'] == 'M'
print(M)
```

```python
Customer_Transactions['state'] = Customer_Transactions['state'].replace(['New South Wales'],'NSW')
```

```python
#confirming the replacement was successful through condional extraction
NewSouthWales = Customer_Transactions['state'] == 'New South Wales'
print(NewSouthWales)
```

```python
Customer_Transactions['state'] = Customer_Transactions['state'].replace(['Victoria'],'VIC')
```

```python
#confirming the replacement was successful through condional extraction
Victoria = Customer_Transactions['state'] == 'Victoria'
print(Victoria)
```

***How are more of the orders placed ,online or not?***

```python
# grouping the data
Online_orders = Customer_Transactions.groupby(['online_order'])['customer_id'].agg('count')
print(Online_orders)
```

```python
# plotting if the order was online or not to see what instance happens alot
Online_orders.plot(kind= 'bar')
plt.title('Online Orders')
plt.ylabel('No of Orders')
plt.xlabel('Online or Not')
plt.show()

#0.0 means the order was not done online,false, while 1.0 means it was done online,true with this their is no big difference though online orders tend to be more than the orthers with a slight difference of 21 orders. 
```
